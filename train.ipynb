```python
X = X.reshape(X.shape[0],5512,1)
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import classification_report
from sklearn.cross_validation import StratifiedKFold, cross_val_score
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import backend as K
import numpy as np
from keras.layers import *
import keras
from keras.models import Model

def create_model():
    sequence_input = Input(shape=(5512,1))
    l_conv1 = Conv1D(128, 5, activation='relu')(sequence_input)
    l_pool1 = MaxPooling1D(5)(l_conv1)
    l_conv2 = Conv1D(128, 5, activation='relu')(l_pool1)
    l_pool2 = MaxPooling1D(5)(l_conv2)
    #l_drop2 = Dropout(0.4)(l_pool2)
    l_conv3 = Conv1D(128, 5, activation='relu')(l_pool2)
    l_pool3 = MaxPooling1D(5)(l_conv3)  # global max pooling
    l_flat = Flatten()(l_pool3)

    l_dense = Dense(128, activation='relu')(l_flat)
    preds = Dense(2, activation='softmax')(l_dense)

    model = Model(inputs = sequence_input, outputs = preds)
    model.compile(loss='categorical_crossentropy',
             optimizer='rmsprop',
             metrics=['acc'])

    return model
```

```python
p = 0
r = 0
f = 0
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    #X_train = X_train.reshape(X_train.shape[0],5512,1)
    #X_test = X_test.reshape(X_test.shape[0],5512,1)
    clf = KerasClassifier(build_fn=create_model, epochs=12, batch_size=26)
    #clf = RF(n_estimators=10, random_state=2)
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_test)
    #precision_recall_fscore_support(y_true, y_pred, average='macro')
    a=precision_recall_fscore_support(y_test, y_pred,average='macro')
    print(a)
    p = p + a[0]
    r = r + a[1]
    f = f + a[2]
print('precision:',p/10)
print('recall:',r/10)
print('f1-score:',f/10)
```

```python
p = 0
r = 0
f = 0
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    X_train = X_train.reshape(X_train.shape[0],5512,1)
    X_test = X_test.reshape(X_test.shape[0],5512,1)
    y_train = keras.utils.to_categorical(y_train, 2)
    y_test = keras.utils.to_categorical(y_test, 2)
    sequence_input = Input(shape=(5512,1))
    l_conv1 = Conv1D(128, 5, activation='relu')(sequence_input)
    l_pool1 = MaxPooling1D(5)(l_conv1)
    l_conv2 = Conv1D(128, 5, activation='relu')(l_pool1)
    l_pool2 = MaxPooling1D(5)(l_conv2)
    #l_drop2 = Dropout(0.4)(l_pool2)
    l_conv3 = Conv1D(128, 5, activation='relu')(l_pool2)
    l_pool3 = MaxPooling1D(5)(l_conv3)  # global max pooling
    att_layer = AttLayer()
    l_att=att_layer(l_pool3)
    l_flat = Flatten()(l_att)
    l_dense = Dense(128*2, activation='relu')(l_flat)

    preds = Dense(2, activation='softmax')(l_dense)
    model = Model(sequence_input, preds)
    model.compile(loss='categorical_crossentropy',
                optimizer='rmsprop',
                metrics=['accuracy'])
    model.fit(X_train, y_train,
          batch_size=52,
          epochs=12,
       #   verbose=1,
         )
    #clf = KerasClassifier(build_fn=create_model1, epochs=10, batch_size=52,verbose=1)
    #clf = RF(n_estimators=10, random_state=2)
    #clf.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    yy_pred = [np.argmax(i) for i in y_pred]
    yy_true = [np.argmax(i) for i in y_test]
    #precision_recall_fscore_support(y_true, y_pred, average='macro')
    a=precision_recall_fscore_support(yy_true,yy_pred,average='macro')
    print(a)
    p = p + a[0]
    r = r + a[1]
    f = f + a[2]
print('precision:',p/10)
print('recall:',r/10)
print('f1-score:',f/10)
```
